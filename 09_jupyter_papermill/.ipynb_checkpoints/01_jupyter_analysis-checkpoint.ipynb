```python
# Imports

# Allows us to make markdown programmatically
from IPython.display import Markdown as MD
from IPython.display import HTML

# Core Packages
import pandas as pd
import numpy as np

# Formatters
from mizani.formatters import label_dollar

# Custom Package
from my_panda_extensions.database import read_forecast_from_database
from my_panda_extensions.forecasting import plot_forecast
```

```python
# Pull Data from Database
all_forecast_df = read_forecast_from_database()
all_forecast_df
```

```python
# Inspect Unique ID's
all_forecast_df['id'].unique()
```

```python
# Filter columns to ones we want
ids = ['Total Revenue', "Category 1: Road", "Category 1: Mountain"]
_filter = all_forecast_df['id'].isin(ids)
forecast_filtered_df = all_forecast_df[_filter]
forecast_filtered_df
```

## Performance Summary

```python
# Pivot Table Summary
usd = label_dollar(big_mark = ",")

forecast_filtered_df.loc[:, ['id', 'date', 'value']]\
    .dropna() \
    .groupby(['id', 
              pd.Grouper(key = 'date', freq = 'Y',origin = 'start'),]) \
    .agg(np.sum) \
    .reset_index() \
    .assign(year = lambda x: x['date'].dt.year) \
    .pivot_table(
           values = "value",
           columns = "year",
           index = "id",
           aggfunc = np.sum
    )\
    .style \
    .background_gradient(cmap = "Blues", axis =1) \
    .format(lambda x: usd([x])[0])
```
